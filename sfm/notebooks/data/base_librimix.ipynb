```python
from sfm.data.base_librimix import BaseLibriMix as DS
from sfm.data.base_datamodule import BaseDatamodule as DM
import yaml
from itertools import islice
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import torch
from IPython.display import Audio, display

N_SPEAKER = 2
CONFIG_PATH = '../../config/'
```

```python
# load default parametrization
base_config_path = CONFIG_PATH + 'base_config.yaml'
with open(base_config_path, 'r') as file:
    config = yaml.safe_load(file)
config['audio_params']['n_speaker'] = N_SPEAKER

# instantiate dataloader
dm = DM(
    dataset=DS,
    batch_size=1, 
    n_workers=0, 
    **config,
)
val_dl = dm.val_dataloader()
```

```python
# load examples
example_idx = 3
example = next(islice(val_dl, example_idx, None))

# display waveform
fig, axs = plt.subplots(ncols=N_SPEAKER, figsize=(4 * N_SPEAKER, 2), sharey=True)
time = torch.arange(*example['audio_samples']) / config['audio_params']['sample_rate']
print('#'*30, f'example {example_idx}', '#'*30)
for spk_idx in range(N_SPEAKER):
    axs[spk_idx].set_title(f'speaker {spk_idx}')
    axs[spk_idx].plot(time, example['clean_td'][0, spk_idx])
    axs[spk_idx].xaxis.set_major_locator(MultipleLocator(1))
    axs[spk_idx].set_xlabel('time in s')
plt.show()

# display audio
for spk_idx in range(N_SPEAKER):
    print(f'speaker {spk_idx}')
    display(Audio(example['clean_td'][0, spk_idx], rate=config['audio_params']['sample_rate']))
```
